```python
import pandas as pd
from kafka import KafkaConsumer
from time import sleep
from json import dumps, loads, dump
import os
from google.cloud import storage
import tempfile
```

```python
consumer = KafkaConsumer('stock_topic',
                         bootstrap_servers=['localhost:9092'],
                         value_deserializer=lambda x: loads(x.decode('utf-8'))
                        )
```

```python
def write_to_gcs(bucket_name, json_data, destination_blob_name):
    keyfile_path = os.environ.get('GOOGLE_APPLICATION_CREDENTIALS_PATH')
    if keyfile_path is None:
        raise ValueError("Environment variable GOOGLE_APPLICATION_CREDENTIALS_PATH is not set.")

    client = storage.Client.from_service_account_json(keyfile_path)
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    # blob.upload_from_filename(source_file_path)
    blob.upload_from_string(dumps(json_data), content_type="application/json")

```

```python
for count, i in enumerate(consumer):
    print(f"Received the data from Producer- {i.value}")
    file_name = 'stock_data_{}.json'.format(count)
    write_to_gcs(bucket_name='stock_kafka_data', json_data=i.value, destination_blob_name=file_name)
```
